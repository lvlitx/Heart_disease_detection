```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# In the above case 1 means defective heart and 0 means good heart
#drop the target column that is the target variable
X = heart_data.drop(columns='target' , axis=1)
```

```python
#store the target variable in the Y variable
Y = heart_data['target']
```

```python
print(X.shape)
print(X)
```

```python
print(Y.shape)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X , Y , test_size=0.2 , stratify=Y , random_state=2)
```

```python
print( X.shape , X_train.shape , X_test.shape)
```

```python
# model training
model = LogisticRegression()
```

```python
model.fit(X_train ,Y_train)
```

```python
#model evatluation and accuracy score
trained_predicn = model.predict(X_train)
```

```python
# checking the training accuracy
training_data_accuracy = accuracy_score(trained_predicn, Y_train)
```

```python
print(training_data_accuracy)
```

```python
#predicting the testing data with the test data
test_prediction = model.predict(X_test)
```

```python
#accuaracy on testing data
testing_data_predict = accuracy_score(test_prediction ,Y_test)
```

```python
print("The accuracy of the data is :" ,testing_data_predict )
```
